```python
import pandas as pd
```

```python
input_file = '/gpfs/home/ptseng/Torkamani_Projects/20210830_PCa_UKBB/fuma/fuma_snp_output/snps.txt'
```

```python
data = pd.read_csv(input_file,sep='\t')
data = data.loc[data['gwasP'].notna()]
data = data.drop_duplicates('rsID')
data
```

```python
snp2gene = data[['rsID','nearestGene']]
snp2gene = snp2gene.rename(columns={
    'rsID':'SNP',
    'nearestGene':'gene'
})
snp2gene
```

```python
snp2gene.to_csv('../snp2gene.txt',sep=' ',index=False)
```

```python
snp_chr_pos_p = data[['rsID','chr','pos','gwasP']]
snp_chr_pos_p = snp_chr_pos_p.rename(columns={
    'rsID':'SNP',
    'gwasP':'p'
})
snp_chr_pos_p
```

```python
snp_chr_pos_p.to_csv('../snp_chr_pos_p.txt',sep=' ',index=False)
```
